```python
# read datasets/superblue1.nodes

file = open("datasets/superblue1/superblue1.nodes", "r")

# ignore the first 4 lines
for _ in range(4):
    file.readline()

# read line "NumNodes      :  847441"
line = file.readline()
# split the line into parts
parts = line.split()
# get the number of nodes
num_nodes = int(parts[2])

# read line "NumTerminals  :  82339"
line = file.readline()
# split the line into parts
parts = line.split()
# get the number of terminals
num_terminals = int(parts[2])

# skip an empty line
file.readline()

nodes = []
nodes_width = []
nodes_height = []

# read num_nodes - num_terminals lines, each line belike "             o0             5             9"
for _ in range(num_nodes - num_terminals):
    line = file.readline()
    # split the line into parts
    parts = line.split()
    # get the node name, width, and height
    node_name = parts[0]
    node_width = int(parts[1])
    node_height = int(parts[2])
    # append to the list
    nodes.append(node_name)
    nodes_width.append(node_width)
    nodes_height.append(node_height)

terminals = []
terminals_width = []
terminals_height = []

# log scale the node widths and heights
# import numpy as np
# nodes_width = np.log10(nodes_width)
# nodes_height = np.log10(nodes_height)

# find the max width and height
max_width = max(nodes_width)
max_height = max(nodes_height)
print("Max width: ", max_width)
print("Max height: ", max_height)

bound_width = 30
bound_height = 10

nodes_small = []
nodes_small_width = []
nodes_small_height = []
nodes_large = []
nodes_large_width = []
nodes_large_height = []

# split the nodes into small and large
for i in range(num_nodes):
    if nodes_width[i] < bound_width and nodes_height[i] < bound_height:
        nodes_small.append(nodes[i])
        nodes_small_width.append(nodes_width[i])
        nodes_small_height.append(nodes_height[i])
    else:
        nodes_large.append(nodes[i])
        nodes_large_width.append(nodes_width[i])
        nodes_large_height.append(nodes_height[i])
# print the number of small and large nodes
print("Number of small nodes: ", len(nodes_small))
print("Number of large nodes: ", len(nodes_large))

import matplotlib.pyplot as plt
plt.hist(nodes_small_width, bins=100)
plt.title("Small Node Widths")
plt.xlabel("Width")
plt.ylabel("Frequency")
plt.show()
plt.hist(nodes_small_height, bins=100)
plt.title("Small Node Heights")
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.show()

plt.hist(nodes_large_width, bins=100)
plt.title("Large Node Widths")
plt.xlabel("Width")
plt.ylabel("Frequency")
plt.show()
plt.hist(nodes_large_height, bins=100)
plt.title("Large Node Heights")
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.show()

# # draw the frequency of the node widths and heights
# plt.hist2d(nodes_large_width, nodes_large_height, bins=100)
# plt.title("Large Node Widths and Heights")
# plt.xlabel("Width")
# plt.ylabel("Height")
# plt.colorbar()
# plt.show()


# # draw the frequency of the node widths
# import matplotlib.pyplot as plt
# plt.hist(nodes_width, bins=100)
# plt.title("Node Widths")
# plt.xlabel("Width")
# plt.ylabel("Frequency")
# plt.show()
# # draw the frequency of the node heights
# plt.hist(nodes_height, bins=100)
# plt.title("Node Heights")
# plt.xlabel("Height")
# plt.ylabel("Frequency")
# plt.show()
# # draw the frequency of the node widths and heights
# plt.hist2d(nodes_width, nodes_height, bins=100)
# plt.title("Node Widths and Heights")


# close the file
file.close()
```
